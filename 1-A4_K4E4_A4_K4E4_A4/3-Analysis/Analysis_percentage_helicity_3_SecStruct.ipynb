```python
df_tot=df_alpha

df_tot['Fraction beta']=df_anti['Fraction anti']+df_para['Fraction para']

df_tot['Alpha helix'] = df_tot['Fraction helicity']*33/31
df_tot['3-10 helix'] = df_310['Fraction 310']*33/31
df_tot['Pi helix'] = df_pi['Fraction pi']*33/31
df_tot['Beta strand'] = df_tot['Fraction beta']*33/31
df_tot['Turn'] = df_turn['Fraction turn']*33/31
df_tot['Bend'] = df_bend['Fraction bend']*33/31
df_tot['Coil'] = 1-(df_tot['Beta strand']+df_tot['Alpha helix']+df_tot['3-10 helix']+df_tot['Turn']+
                   df_tot['Bend']+df_tot['Pi helix'])
df_tot['Other'] = 1-(df_tot['Beta strand']+df_tot['Alpha helix'] )

df_tot=df_tot.drop(['Frame','Fraction helicity','Fraction beta'], axis=1)

df_tot.head()
```

```python
root_path = '../2-Run1'  # The path to the directory containing the systems

igbs = ['igb1', 'igb2', 'igb5', 'igb8']  # Each of the systems as named in their directory

ffs = ['ff94','ff96','ff98','ff99', 'ff99SB', 'ff99SBildn', 'ff99SBnmr', 'ff03.r1', 'ff14SB', \
       'ff14SBonlysc','ff14ipq','fb15','ff15ipq']  # The name of the state directories

filename_pattern = 'helicity.dat'  # The glob to match the data files to read

dfs = {}  # Mapping of system name to list of state DataFrames

igbff_all = []
mean=[]
std=[]
igbff=[]
igbl=[]
ffl=[]
for ff in ffs:
    for igb in igbs:
        direc=igb + '_'+ ff
        file_glob = os.path.join(root_path, direc, filename_pattern)
        files = glob.glob(file_glob)
        for f in files:
            df = pd.read_csv(f, sep=r'\s+', skiprows=1,usecols=[1],names=['Fraction helicity'])
            df['helicity'] = df['Fraction helicity']*33/31*100
            df=df.drop('Fraction helicity',axis=1 )
            mean.append(df.mean()[0])
            std.append(df.std()[0])
            igbff.append('{0} {1}'.format(ff, igb))
            igbl.append(igb)
            ffl.append(ff)
            
mean_series = pd.Series(mean) 
std_series = pd.Series(std)
igbff_series = pd.Series(igbff)
igb_series = pd.Series(igbl)
ff_series = pd.Series(ffl)
  
frame = { 'Average helicity (%)': mean_series, 'Standard deviation': std_series,
        'igbff':igbff_series, 'igb':igb_series, 'ff':ff_series} 
  
df_helicity = pd.DataFrame(frame)             

df_helicity.head()
```

```python
sns.set(context='paper', style='whitegrid', palette="pastel", font='sans-serif',\
        font_scale=1.4, rc={'grid.linestyle': ''}, color_codes=False)

f, ax = plt.subplots(figsize=(10, 5.2))

ax = sns.scatterplot(x="ff", y="igb",
                     hue="Standard deviation", size="Average helicity (%)",
                     palette='GnBu_r', sizes=(5, 1000),
                     data=df_helicity)
sns.despine(left=True, bottom= True)
plt.legend(bbox_to_anchor=(1.1, 1), loc=2, borderaxespad=0., frameon=False, ncol=1, labelspacing=1.1,)
#ax.set_xticklabels(rotation=90)
plt.xticks(rotation=90)


ax.set_xlabel('Force field')
ax.set_ylabel('Implicit solvent')

plt.tight_layout()
plt.savefig('./bubbleplot1.png', dpi=400)
```
