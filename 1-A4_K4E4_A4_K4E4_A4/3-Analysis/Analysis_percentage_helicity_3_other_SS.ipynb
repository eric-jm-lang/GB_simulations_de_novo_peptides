```python
df_alpha = pd.read_csv('../2-Run1/igb8_ff14SB/helicity.dat', sep=r'\s+', skiprows=1, names=['Frame', 'Fraction helicity'])
df_anti = pd.read_csv('../2-Run1/igb8_ff14SB/anti.dat', sep=r'\s+', skiprows=1, names=['Frame', 'Fraction anti'])
df_para = pd.read_csv('../2-Run1/igb8_ff14SB/para.dat', sep=r'\s+', skiprows=1, names=['Frame', 'Fraction para'])
```

```python
df_tot=df_alpha
df_tot['Fraction beta']=df_anti['Fraction anti']+df_para['Fraction para']
df_tot['Fraction other']=1-(df_tot['Fraction helicity']+df_tot['Fraction beta'])
df_tot['Helicity (%)'] = df_tot['Fraction helicity']*100*33/31
df_tot['Beta (%)'] = df_tot['Fraction beta']*100*33/31
df_tot['Other (%)'] = 100-(df_tot['Beta (%)']+df_tot['Helicity (%)'] )
df_tot=df_tot.drop(['Frame','Fraction helicity','Fraction beta','Fraction other'], axis=1)
```
